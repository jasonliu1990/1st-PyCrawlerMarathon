## 作業目標

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

## 1. 取出今天所有的新聞

```python
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html.parser")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
browser.quit()
```

## 2. 取出現在時間兩小時內的新聞

```python
import datetime
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    t = d.find(class_="date").text[11:16]
    hh = str(t[:2])
    mm = str(t[3:])
    hhmm =  int(hh + mm)
    target_hh = str(datetime.datetime.now().hour - 2)
    target_mm = str(datetime.datetime.now().minute)
    if len(target_mm) == 1:
        target_mm = '0' + target_mm
    target_hhmm = int(target_hh + target_mm)
    if hhmm >= target_hhmm:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
browser = webdriver.Chrome(executable_path='chromedriver')
# 三天前
browser.get("https://www.ettoday.net/news/news-list-2019-12-26-0.htm")
```

```python
while True:
    
    browser.execute_script("window.scrollTo(0, 50000);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html.parser")
    for d in soup.find(class_="part_list_2").find_all('h3'):
        t = d.find(class_="date").text[11:16]
        hh = str(t[:2])
       
    if hh == '14':
        break
    time.sleep(1)
    
for d in soup.find(class_="part_list_2").find_all('h3'):
        t = d.find(class_="date").text[11:16]
        hh = str(t[:2])
        if int(hh) >= 15 and int(hh) <= 17:
            print(d.find(class_="date").text, d.find_all('a')[-1].text)
    
```

```python
browser.quit()
```
