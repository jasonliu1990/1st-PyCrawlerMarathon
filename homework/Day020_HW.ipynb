```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
import pandas as pd
```

```python
check_point = 1
data = pd.DataFrame()
for i in range(1, len(table.find_all('tr'))):
    if i == check_point:
        date_list = []
        values_list = []
        col = table.find_all('tr')[i].find_all('td')[0].text
        fst_date = table.find_all('tr')[i].find_all('td')[2].text
        fst_value = table.find_all('tr')[i].find_all('td')[3].text
        date_list.append(fst_date)
        values_list.append(fst_value)
        check_point += 11      
    else:
        tmp_date = table.find_all('tr')[i].find_all('td')[0].text
        tmp_value = table.find_all('tr')[i].find_all('td')[1].text        
        date_list.append(tmp_date)
        values_list.append(tmp_value)
    
    if i % 11 == 0:
        tmp_df = pd.DataFrame({col: values_list}, index=date_list)
        data = pd.concat([data, tmp_df], axis=1)
```
